```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import streamlit as st
```

```python
df = pd.read_csv("IPL.csv")
df.head()
```

Toss Winner Vs Match Winner

```python
toss_winner = df['toss_winner'].value_counts()
toss_winner
sns.barplot(y = toss_winner.index,x = toss_winner.values)
plt.title("Toss Winners")
plt.xlabel("TEAM Name")
plt.ylabel("Toss")

```

```python
match_winner = df['match_winner'].value_counts()
sns.barplot(y = match_winner.index, x = match_winner.values)
plt.title("Match Winners")
plt.xlabel("Matches")
plt.ylabel("Team")
```

```python
df.head()
identify = df[df['toss_winner']==df['match_winner']].count()
team_who_won_toss_also_won_match = identify['match_winner']
team_who_won_toss_also_won_match
percentage = (team_who_won_toss_also_won_match *100 )/ df.shape[0]
percentage
print(f"There are {percentage}% chance team who won the toss also won the match.")
```

```python
df.head()
avg_score_by_venue = df.groupby('venue')['first_ings_score'].mean()
approximate_avg_score = avg_score_by_venue.values.round()
sns.barplot(y = avg_score_by_venue.index,x = approximate_avg_score)
plt.title("Average First Ining Score")
plt.xlabel("Score")
plt.ylabel("Venue")
```

```python
df.head()
avg_score_by_venue = df.groupby('venue')['second_ings_score'].mean()
approximate_avg_score = avg_score_by_venue.values.round()
sns.barplot(y = avg_score_by_venue.index,x = approximate_avg_score)
plt.title("Average Second Ining Score")
plt.xlabel("Score")
plt.ylabel("Venue")
```

```python
df.head()
top_scorer = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(5)
sns.barplot(y = top_scorer.index,x = top_scorer.values)
plt.title("Top 5 Player score")
plt.xlabel("Runs")
plt.ylabel("Player Name")
```
