```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
def bar_plot(x,y,title,xlabel,ylabel):
    sns.barplot(x=x,y=y,palette='rainbow',hue=y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python
df = pd.read_csv("CWC2023.csv")
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.shape
```

```python
venue = df['Stadium'].unique()
venue
```

```python
df.head()
toss_winner = df['Toss Winner'].value_counts()
toss_winner
bar_plot(toss_winner.values,toss_winner.index,"Toss Winner","Toss","Team")
```

```python
df.head()
venue_by_mean = df.groupby('Stadium')[['Score A','Score B']].mean()
venue_by_mean.plot(kind='barh',figsize=(12,8))
plt.title("Average score of Team A vs Team B")
plt.xlabel("Score")
```

```python
df.head()
```

```python
match_played_venue = df['Stadium'].value_counts()
match_played_venue
plt.pie(match_played_venue,labels=match_played_venue.index)
plt.title("Match Played in each venue")
plt.show()
```

```python
df.head()
```

```python
toss_decision = df['Toss Decision'].value_counts()
toss_decision
plt.pie(toss_decision,labels=toss_decision.index)
plt.title("Toss Decision")
plt.show()
```

```python
df.head()
```

```python
unique_team = df['Team A'].unique()
unique_team
boundaries = df.groupby('Stadium')[['Boundaries A','Boundaries B']].sum()
boundaries.plot(kind='barh',figsize=(12,8))
plt.title("Boundaries in each Venue")
plt.xlabel("Boundaries")
plt.ylabel("Venue")
plt.legend(['Team A','Team B'])
plt.show()
```

```python
df.head()
winning_team = df['Wining Team'].value_counts()
winning_team
bar_plot(winning_team.values,winning_team.index,"Total Win by a Team","Number","Team Name")
```

```python
df.head()
man_of_match = df['Man of the Match'].value_counts().sort_values(ascending=False).head()
man_of_match
bar_plot(man_of_match.values,man_of_match.index,"Top 5 MOM Player","Number","Name")
```
